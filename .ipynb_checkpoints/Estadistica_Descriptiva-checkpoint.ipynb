### Medidas de tendencia central

- son una manera de resumir informacion

## Ventajas

- Obtener informacion valiosa

## Desventajas

- Hay informaciones que no se consiguen entender con una sola medida

## Media

- promedio

- desventaja, es suceptible a valores atipicos, se ve afectada cuando existen valores distantes de los comunes que hay en la muestra

$$\{ x_{1}, x_{2}, ..., x_{n} \} = \text{dataset}$$

$$\text{media} = \bar{x} = \frac{1}{N}(x_{1} + x_{2} + ... + x_{n}) = \frac{1}{N} \sum_{i=1}^{n} x_{i}$$

## Mediana 

- dato central, esta justamente balanceado, no es lo mismo que la media

$$\text{mediana} = \text{median}(x) = 
\begin{cases} 
\frac{x_{\frac{n}{2}} + x_{\frac{n}{2}+1}}{2}, & \text{N par} \\ 
x_{[ \,\frac{N+1}{2}] \,}, & \text{N impar} 
\end{cases}$$

## Moda

- Valor que mas se repite, el valor mas comun

$$\text{moda} = x_{x} \text{ donde} \text{ Freq}(x_k) = \text{max }(\text{Freq}(x_i))$$

## Medidas de dispersion

- complemento para medidas de tendencia central

- las mas famosas son el rango, rango intercuartil y desviacion estandard

- miden que tan dispersos estan los datos

### Rango

- diferencia entre valor maximo y valor minimo

- nos da la maxima medida de la dispersion del conjunto


<img src="Screenshot from 2022-10-14 11-11-06.png" alt="MarineGEO circle logo" style="height: 500:px; width:500px;"/>

### Rango intercuartil (iqr)

- es la distancia entre q1 e q2

- para entenderlo mejor los datos deben ser divididos en 4 subdivisiones homogeneas
- partimos del segundo quartil q2,  que representa la mediana y que parte a la mitad los datos
- luego, el primer quartil, q1, valor entre la mediana y el valor minimo
- tercer quartil, q3, valor entre la mediana y el valor maximo
- un boxplot ofrece una representacion simplificada de la distribucion de los datos en referencia a la mediana

### Desviacion estandar

- medida de distribucion mas utilizada en el area de estadistica descriptiva, debe considerarse cuando tenemos una distribucion de datos simetrica o normal

- el primer concepto asociado a la desviacion estandar es la varianza, que esta dada por la suma de las diferencias de todas las distancias cuadraticas respecto del promedio elevadas al cuadrado

$$\text{Varianza}=\sigma^2=\frac{1}{n} \sum_{i=1}^{n} (x_{i}-\mu)^2$$

$$\text{Desviacion standad}=\sigma=\sqrt{\frac{1}{n-1} \sum_{i=1}^{n} (x_{i}-\mu)^2}$$


- podemos obtener la desviacion standard de la poblacion o de una muestra, cuando obtenemos la distribucion standad de una muestra es comun encontrar una correcion como se ve acontinuacion


$$\text{Desviacion standad}=\sigma=\sqrt{\frac{1}{n-1} \sum_{i=1}^{n} (x_{i}-\mu)^2}$$

- cuando tomo los datos entre $\mu-3\sigma$ y $\mu+3\sigma$ estoy considerando el 99.72% de los datos

- los outliers o datos anomalos son datos no comunes que se alejan de la media por lo menos $3\sigma$

- El metodo de deteccion de outliers con el rango intercuartil nos permite incluir los datos significativos de estos datos anomalos, aproximandose a $\mu-3\sigma$ y $\mu+3\sigma$: 

$$\text{min}=Q1-1.5\text{IQR}$$
$$\text{max}=Q3+1.5\text{IQR}$$

- Para una distribucion de datos asimetrica usamos funciones que se ajusten a la asimetria de la funcion

$$\text{min}=Q1-1.5f(\text{IQR})$$
$$\text{max}=Q3+1.5g(\text{IQR})$$

### Coeficiente de variabilidad

- division de la desviacion estandar entre la media

## Pipelines de procesamiento de datos numericos

- uso de estadistica descriptiva para pre procesamiento de datos

### Escalamiento lineal

- es importante normalizar los datos  ya que los optimizadores de los modelos de ML son optimos en la medida en que todos los atributos tengan las mismas dimensiones, numeros grandes disminuyen la convergencia

- lo ideal es tener datos entre -1 y 1

- escalamientos lineares funcionan cuando la data esta distribuida de forma simetrica

### Escalamiento min-max

$$x\longrightarrow x_{s}$$

$$x_{s}=\frac{2x-min-max}{max-min}$$

### Escalamiento clipping

- no muy recomendado

- lleva los datos por fuera de la distribucion a valores limite determinados

<img src='Screenshot from 2022-10-19 18-11-51.png' alt="MarineGEO circle logo" style="height: 500:px; width:500px;"/>

<img src='Screenshot from 2022-10-19 18-12-08.png' alt="MarineGEO circle logo" style="height: 500:px; width:500px;"/>

- no es muy recomendable ya que modifica la realidad de los datos

### z-score

- esta basado en el promedio y desviacion standard
- basandonos en el hecho de que los datos tienen una distribucion simetrica, se obtiene la media y desviacion standard

$$x_{s}=\frac{x-\mu}{\sigma}$$

- dada la transformacion, la distribucion sigue siendo gausiana con media 0 y desviacion standard de 1

- pero, que hacer cuando la distribucion de los datos no es simetrica?

### Transformacion no lineal

- la solucion es transformar los datos de forma simetrica

- funciones no lineales como: logaritmos, sigmoides, polinomiales, etc

- es un paso anterior al escalamiento lineal

- ex.

$$x_{s}=tanh(x)$$

 - existen muchas transformaciones que pueden ser aplicadas a los datos para hacerlos simetricos

```python
import timeit
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = datasets.load_diabetes(return_X_y=True)
raw = X[:, None, 2]
```

- primero alistamos el escalamiento
- para escalamiento min-max:

```python
print(np.median(raw), np.std(raw))
```

```python
max_raw = max(raw)
min_raw = min(raw)
scaled_max_min = (2*raw-max_raw-min_raw)/(max_raw-min_raw)
scaled_z_score = (raw-np.median(raw))/np.std(raw)

fig, axs = plt.subplots(3, 1, sharex=True)

axs[0].hist(raw)
axs[1].hist(scaled_max_min)
axs[2].hist(scaled_z_score)
```

```python
# modelos de entrenamiento sin escala y con escala

def train_raw():
    linear_model.LinearRegression().fit(raw, y)
    
def train_scaled_max_min():
    linear_model.LinearRegression().fit(scaled_max_min, y)
    
def train_scaled_z_score():
    linear_model.LinearRegression().fit(scaled_z_score, y)
```

```python
raw_time = timeit.timeit(train_raw, number=100)
scaled_max_min_time = timeit.timeit(train_scaled_max_min, number=100)
scaled_z_score_time = timeit.timeit(train_scaled_z_score, number=100)
print(f'train raw: {raw_time}')
print(f'train scaled: {scaled_max_min_time}')
print(f'scaled_z_score: {scaled_z_score_time}')
```

- El output muestra una disminucion en el tiempo para el entrenamiento con datos escalados
- como configuracion standard:

  max-min scalling es mejor para datos uniformemente distribuidos
  
  z-score scalling es mejor para datos distribuidos 'normalmente'(forma de campana de gauss)

## Correlaciones

- si dos variables siguen la misma tendencia significa que estan coorelacionadas con lo que no tiene sentido incluirlas en el mismo modelo, ya que ambas variables estarian aportando la misma informacion

- es por eso que resulta importante hacer buscar las correlaciones

- para medir cuan relacionadas estan dos variables es preciso el concepto de covarianza, que suma las multiplicaciones de las varianzas de las dos variables que se desean entender

$$\text{Covarianza}(x, y)=\frac{1}{n} \sum_{i=1}^{n} (x_{i}-\mu)(y_{i}-\mu)$$

- debido a que las variables pueden estar en diferentes unidades o magnitudes se hace necesario dividir las variaciones de cada una de las variables entre sus respectivas desviaciones standard

- el coeficiente de correlacion realiza esta medicion

$$\rho=\frac{Covarianza}{\sigma_{x}\sigma_{y}}$$

- cuando el coeficiente de correlacion es alto las variables tienen una correlacion alta, y cuando es bajo su correlacion es baja

- si el valor es negativo significa que la relacion entre variables es inversa, esto significa que si una variable aumenta la otra disminuye

- correlacion no implica causacion

## Matriz de covarianza

- al tener un dataset con varias variables, la idea es calcular todas las posibles covarianzas para determinar las relaciones entre esas variables

```python
# representacion analitica

scaler = StandardScaler()
scaled = scaler.fit_transform(
    iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
)

# debe sin embargo transponerse la matriz para poder obtener la matriz de correlaciones ya que el formato con el que viene no lo consigue

scaled.T
```

```python
# para hayar la matriz de covarianza utilizamos np.cov() para numpy o 

covariance_matrix=np.cov(scaled.T)
covariance_matrix
```

```python
plt.figure(figsize=(10,10))
sns.set(font_scale=1.5)
hm=sns.heatmap(covariance_matrix,
               cbar=True,
               annot=True,
               square=True,
               fmt='.2f',
               annot_kws={'size':12},
               yticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
               xticklabels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
```

- esta informacion sera util para seleccionar las variables para el modelo de machine learning en una futura reduccion de variables

```python
# el producto entre matrices se calcula usando la libreria numpy
# recordando que solo pueden ser multiplicadas cuando tienen la misma dimension

A=np.array([[2,4],[-1,2]])
B=np.array([[3,6],[7,-2]])

np.matmul(A, B)
```

```python
# para calcular la matriz inversa usamos:

A_inversa=np.linalg.inv(A)

print(A)
print(A_inversa)
```

```python
# para verificar que el resultado es correcto podemos multiplicar la matriz por su inversa para obtener 1 como resultado

np.matmul(A, A_inversa)
```

## Principal component analysis (PCA)

- permite representar la mayor cantidad de varianza en terminos de la menor cantidad de variables

### Reduccion de dimensionalidad de datos con PCA

- la idea es determinar el menor numero de variables que defina el data set

```python
# para obtener los valores propios o "eigen" values usamos la siguiente funcion de numpy

eigen_values, eigen_vectors=np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
# cada vector representa una direccion principal a lo largo de la cual
#capturaremos la varianza de los datos originales

variance_explained=[]
for i in eigen_values:
    variance_explained.append((i/sum(eigen_values))*100)
    
print(variance_explained)
```

- el primer vector expresa el 72% de la varianza total de los datos
- el segundo vector expresa el 23%
- el tercero apenas el 4%
- y el ultimo el 0.5%
- indicando que la tercera y cuarta componentes pueden ser despreciables del analisis
- para reducir la dimensionalidad de los datos podemos usar sklearn.decomposition


```python
from sklearn.decomposition import PCA

pca=PCA(n_components=2) # n_components dice el numero de componentes al cual desea ser reducida la informacion
pca.fit(scaled)
```

```python
pca.explained_variance_ratio_ # explica los valores que mas varianza capturan
```

- la idea es obtener nuevas variables de la combinacion de las antiguas

```python
reduced_scaled_var=pca.transform(scaled)
```

```python
reduced_scaled_var
```

```python
iris['pca_1']=reduced_scaled_var[:,0]
iris['pca_2']=reduced_scaled_var[:,1]

iris
```

```python
sns.jointplot(iris['pca_1'], iris['pca_2'], hue=iris['species'])
```

- pca representa la culminacion del pipeline para el curso de estadistica descriptiva
- ahora los datos estan listos para ser entregados al modelo predictiva
- sin embargo esta no es la unica tecnica de extraccion de caracteristicas
